```python
# importing required libraries
import matplotlib.pyplot as plt    
import seaborn as sb              
import pandas as pd

# reading the CSV file and storing it to dataframe 'df_csv'
df_csv = pd.read_csv('../data-clean/StudentsPerformance_Clean.csv')

# adding total sccore and average score columns
df_csv['total score'] = df_csv['math score'] + df_csv['reading score'] + df_csv['writing score']
df_csv['average score'] = df_csv['total score'] / 3
df_csv.head()

#displaying dataframe
df_csv
```

```python
# VISUALIZATION - 1

plt.rcParams['figure.figsize'] = (30, 12)

# Subplot between genders
plt.subplot(1, 4, 1)
ax =sb.histplot(df_csv, x = 'gender', color='red')

# Subplot between race/ethnicity
plt.subplot(1, 4, 2)
ax = sb.histplot(df_csv, x = 'race/ethnicity', color='Green')

# Subplot between parental level of education
plt.subplot(1, 4, 3)
ax = sb.histplot(df_csv, x = 'parental level of education',color='blue')
plt.setp(ax.get_xticklabels(), rotation=20, horizontalalignment='right')

# Subplot between lunch system
plt.subplot(1, 4, 4)
ax = sb.histplot(df_csv, x = 'lunch', color='yellow')

plt.grid()
plt.show()
```

1. We are able to conclude from the foregoing that data is not biased towards either gender.Â since the data almost equalizes the number of samples for men and women.

2. In this data sample, almost 30% of the students are from ethnic group C which is four times the size ofÂ group A.

3. Students with master's or bachelor's degree-holding parents perform worse than in the otherÂ cases combined.

4. Standard lunches are more beneficial to pupils than free or reduced meals.

```python
# VISUALIZATION - 2

plt.subplots(1,3,figsize=(20,20))

# subplot between average score and lunch system
plt.subplot(1,3,1)
ax =sb.histplot(data=df_csv,x='average score',hue='lunch')

# subplot between average score and lunch system for males
plt.subplot(1,3,2)
ax =sb.histplot(data=df_csv[df_csv.gender=='male'],x='average score',hue='lunch')

# subplot between average score and lunch system for females
plt.subplot(1,3,3)
ax =sb.histplot(data=df_csv[df_csv.gender=='female'],x='average score',hue='lunch')

plt.show()
```

350 students receive free or discounted meals.All other meals are typical ones. Therefore, it follows that regardless of gender, a normal meal has a higher average score than a free or reduced lunch.

```python
# VISUALIZATION - 3

# bar chart between race/ethnicity, math score and gender
sb.barplot(data=df_csv,x='race/ethnicity',y='math score',hue='gender')
```

Race and ethnicity are shown against the gender-based average math score in the graph above. We may infer from the graph above that men have much higher averages than women, and this is true regardless of race. however, group E members of both sexes have math averages that are greater than those of all other races.

```python
# VISUALIZATION - 4

# categorical plot between gender, reading score and test preparation course
sb.catplot(data=df_csv,x='gender',y='reading score',kind='box',col='test preparation course', whis=[0,100])

plt.show()
```

According to the following graph, students who finished a course had higher reading average scores than those who had not finished a course or taken any courses, regardless of gender. and overall, female pupils outperformed male students in reading scores. This is totally at odds with math test results.

```python
# VISUALIZATION - 5

# violinplot between parental level of education, average scores and gender
sb.violinplot(df_csv,x="parental level of education",y="average score",hue='gender', split=True)
```

According to the aforementioned graph, guys with master's-degree-holding parents do on average somewhat better than girls. compared to other credentials. and regardless of the educational level of their parents, females perform better than boys overall. Parents with master's degrees tend to have children who do well academically. the opposite need not be the case.
